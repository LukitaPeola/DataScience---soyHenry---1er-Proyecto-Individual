Importo librerías.

```python
import pandas as pd
import numpy as np
import ast
import json
import warnings
import os
warnings.filterwarnings("ignore")
```

```python
# Obtener la ruta actual
ruta_actual = os.getcwd()
```

El archivo no se puede abrir comúnmente, se adecúa para su correcta apertura.

```python
ruta_games = os.path.join(ruta_actual, 'PI MLOps - STEAM', 'output_steam_games.json')
ruta_nueva_games = os.path.join(ruta_actual, 'PI MLOps - STEAM', 'games.json')
```

```python
#games
with open(ruta_games, 'r', encoding='utf-8') as archivo:
    lineas = archivo.readlines()
texto = []
for linea in lineas:
    texto_de_linea = linea.replace('"', '').replace('publisher: ',  '"publisher": "').replace(', genres: ',  '", "genres": "').replace(', app_name:',  '", "app_name": "').replace(', title:',  '", "title": "').replace(', url:',  '", "url": "').replace(', release_date:',  '", "release_date": "').replace(', tags:',  '", "tags": "').replace(', reviews_url:',  '", "reviews_url": "').replace(', specs:',  '", "specs": "').replace(', price:',  '", "price": "').replace(', early_access:',  '", "early_access": "').replace(', id:',  '", "id": "').replace(', developer:',  '", "developer": "').replace('NaN', 'None').replace('}', '"}').replace("\\", "").replace('"}{atch', 'atch').replace(";", ",")
    texto.append(texto_de_linea)

archivo = open(ruta_nueva_games, "w", encoding='utf-8')
archivo.write('[' + '\n')
for i in range(len(texto)-1):
    archivo.write(texto[i] + ',')
archivo.write(texto[len(texto)-1])
archivo.write(']' + '\n')
archivo.close()
```

Creo un DataFrame con lo que creé.

```python
df_games = pd.read_json(ruta_nueva_games, encoding='utf-8')
```

```python
df_games
```

Limpieza de datos.

```python
#Elimino las filas nulos
df_games.replace("None", np.nan, inplace=True)
df_games = df_games.dropna().reset_index(drop=True)
```

```python
df_games
```

```python
#remplazo valores nulos por NaN 
df_games.replace(['None', 'null',''],None, inplace=True)
```

```python
#renombro las columnas
df_games.rename(columns={'app_name': 'name', 'id': 'item_id'}, inplace=True)
```

```python
df_games.info()
```

```python
#Valores nulos de cada columna
df_games[['publisher', 'developer','item_id','price']].isnull().sum()
```

```python
#Dropeo lo que no voy a necesitar y los item_id nulos (por si queda alguno).
df_games.drop(columns=['publisher','title','url','reviews_url','early_access'], inplace=True)
df_games.dropna(subset=['item_id'], inplace=True)
```

```python
df_games.info()
```

```python
# Función segura para evaluar literales
def safe_literal_eval(val):
    try:
        return ast.literal_eval(val)
    except (ValueError, SyntaxError):
        return val

# Convertimos la columna 'genres' a tipo lista, conservando NaN
df_games['genres'] = df_games['genres'].apply(lambda x: safe_literal_eval(x) if isinstance(x, str) else x)

# Expandimos las listas en filas
df_genres = df_games.explode('genres')

# Obtener la lista de géneros únicos, incluyendo los valores nulos
lista_generos = df_genres['genres'].unique()

# Convertimos la columna 'tags' a tipo lista, conservando NaN
df_games['tags'] = df_games['tags'].apply(lambda x: safe_literal_eval(x) if isinstance(x, str) else x)

# Filtramos los tags que están en la lista de géneros y los asignamos a la columna 'genres'
df_games['genres'] = df_games.apply(
    lambda row: [tag for tag in row['tags'] if tag in lista_generos] if isinstance(row['tags'], list) else row['genres'],
    axis=1
)
```

```python
print(df_games.head())
```

```python
#Convertimos la columa release_date a formato fecha
df_games['release_date']=pd.to_datetime(df_games['release_date'], errors='coerce', exact=False)

#Solo nos interesa el año

# Extraer el año y crear una nueva columna 'release_year'
df_games['year'] = df_games['release_date'].dt.year.astype('Int64')

# Eliminar la columna 'release_date'
df_games = df_games.drop(columns=['release_date'])
```

```python
df_games
```

```python
df_games['year'] = df_games['year'].fillna(df_games['year'].median())
```

```python
# Convertir 'price' a tipo numérico, asignar NaN a 'Free To Play'
df_games['price'] = pd.to_numeric(df_games['price'], errors='coerce')

# Reemplazar NaN por 0 en la columna 'price'
df_games['price'].fillna(0, inplace=True)
```

```python
df_games
```

```python
df_games.sort_values(by=['item_id'], inplace=True)
df_games
```

```python
df_games.info()
```

```python
#Intentando transformar una columna apareció un tipo de null que no fue borrado antes
df_games = df_games[df_games['item_id'] != ' None']
```

```python
#Adecúo la columna al mejor tipo de variable para posterior uso.
df_games['price'] = df_games['price'].astype(float)
df_games['item_id'] = df_games['item_id'].astype(int)
```
